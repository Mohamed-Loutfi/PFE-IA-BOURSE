```python
import os
import urllib3
import requests
from datetime import datetime, timedelta
import time

#Configuration
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
base_url = "https://media.casablanca-bourse.com/sites/default/files/es-auto-upload/fr/"
save_folder = "resumes_seance"
start_date = datetime(2025, 1, 1)
end_date = datetime(2025, 12, 31)

#Création dossier
os.makedirs(save_folder, exist_ok=True)

current_date = start_date

while current_date <= end_date:
    
    date_str = current_date.strftime("%Y%m%d")
    filename = f"resume_seance_{date_str}.pdf"
    url = base_url + filename
    save_path = os.path.join(save_folder, filename)

    try:
        response = requests.get(url, timeout=10, verify=False)

        if response.status_code == 200:
            with open(save_path, "wb") as f:
                f.write(response.content)
            print(f"Téléchargé : {filename}")
        else:
            print(f"Non disponible : {filename}")

        # ⏳ Pause pour éviter surcharge serveur
        time.sleep(1)

    except Exception as e:
        print(f"Erreur pour {filename} : {e}")

    current_date += timedelta(days=1)

print("Téléchargement terminé.")
```

```python
import sys
import os
import asyncio
from langchain_google_genai import ChatGoogleGenerativeAI
from browser_use import Agent

if sys.platform == 'win32':
    asyncio.set_event_loop_policy(asyncio.WindowsProactorEventLoopPolicy())

# 1. Configuration
os.environ["GOOGLE_API_KEY"] = ""

# 2. Initialisation du modèle original
llm_original = ChatGoogleGenerativeAI(model="gemini-2.0-flash")

# 3. Création du Wrapper pour contourner les erreurs Pydantic
class GeminiWrapper:
    def __init__(self, llm):
        self.llm = llm
        self.provider = "google"
        self.model_name = llm.model
    
    def __getattr__(self, name):
        return getattr(self.llm, name)

llm = GeminiWrapper(llm_original)

# 4. Création de l'agent
agent = Agent(
    task="Go to https://media.casablanca-bourse.com/sites/default/files/es-auto-upload/fr/resume_seance_20260213.pdf, find the PDF report, and download the files from 01-01-2024 to 31-12-2024 by modifying the url for each file to the /resumes_seance folder",
    llm=llm,
    calculate_cost=False
)

# 5. Exécution
await agent.run()
```
